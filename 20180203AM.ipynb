```python
dic = {}
words = speech.lower.split()
for w in words :
    if w not in dic:
        dic[w]= 1
    else:
        dic[w] = dic[w] + 1
```

```python
import operator
swd = sorted(dic.items(), key = operator.itemgetter(1))
swd
```
