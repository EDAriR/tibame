```python
def page(totalRows,  url, li, headers):
    firstRow = 30
    i = 1
    if totalRows > firstRow:
        firstRow * i
        url = url + 'firstRow=' + str(firstRow) + '&totalRows=' + str(totalRows)
        res = requests.get(url, headers = headers)
        li.extend(res.json().get('data').get('data'))
        i = i + 1
        if (totalRows - firstRow) > firstRow:
            page( firstRow, totalRows,  url, li, headers)
        else:
            return li
    print(len(li))
    
    
page(totalRows,  url, li, headers)
len(li)
```

```python
import time

def houseDeatil(url):
    res = requests.get(url , headers = headers)
    res.text
    soup = BeautifulSoup(res.text, 'lxml')
    price = soup.findAll("div", {"class": "price clearfix"}) #<i>11,500 <b>元/月</b> </i>
    attr = soup.findAll("ul", {"class": "attr"})  #<li>坪數 :  5坪</li><li>樓層 :  4F/5F</li><li>型態 :  公寓</li><li>現況 :  獨立套房</li>

    pprice = soup.select_one('div .price i').text.split(' ')
    aattr = soup.select('ul.attr li')
    aattrList = []
    for tr in  aattr:
        aattrList.append(tr.text.replace("\xa0",""))

    img = soup.select('div ol li a img')
    imgList = []
    for iii in img:
        imgList.append(iii.attrs.get('src').replace('.crop','').replace('125x85','625x425'))

    labelListUnClear = soup.select('ul.labelList li.clearfix')
    labelList = []
    for lll in labelListUnClear:
        labelList.append(lll.text)

    lifeBox = [ele.text for ele in soup.select('.lifeBox p')]

    facilityList = []
    facility = soup.select('.facility li')
    li = [ele for ele in facility ]
    for l in li:
        sp = l.select('span')[0].attrs.get('class')[0]
        if sp != 'no':
            facilityList.append(l.text)

    houseIntro = [ele.text for ele in soup.select('div.houseIntro')]
    time.sleep(3)
    return {
        'labelList':labelList, 'aattrList':aattrList, 'imgList':imgList, 'pprice':pprice, 'lifeBox':lifeBox, 'facilityList':facilityList, 'houseIntro':houseIntro
    }


# houseDeatil('https://rent.591.com.tw/rent-detail-6061063.html')
# houseIntro
# labelList, aattrList, imgList, pprice, lifeBox, facilityList, houseIntro
```
